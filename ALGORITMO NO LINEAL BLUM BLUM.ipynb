## ALGORITMO NO LINEAL BLUM BLUM - ESIS ( Nain Acero)

```python
import numpy as np
import pandas  as pd
import os
import matplotlib.pyplot as plt
```

```python
def no_lineal_blum_blum(n, semilla, m):
    x = []
    for i in range(n):
        numero = (pow(semilla, 2) % m )
        r1 = numero / (m-1)
        x.append(r1)
        semilla = numero
    return x
```

```python
semilla = 17
m = 100
data = no_lineal_blum_blum(50, semilla, m)
print(data)
df = pd.DataFrame({'column':data}) 
df.to_csv('out_algoritmo_no_lineal_blum_blum.txt', index=False, header=['aleatorio'],sep=';',decimal='.')
df.to_csv('out_algoritmo_no_lineal_blum_blum.csv', index=False, header=['aleatorio'],sep=';',decimal='.')
```
