## ALGORITMO CUADRADOS MEDIOS - ESIS ( Nain Acero)

```python
import numpy as np
import pandas  as pd
import os
import matplotlib.pyplot as plt
```

```python
def cuadrados_medios(n, semilla):
    x = []
    tam1 = len(semilla)
    numero1 = int(semilla)
    for i in range(n):
        numero2 = pow(numero1, 2)
        snumero2 = str(numero2)
        tam2 = len(snumero2)
        primerc = (tam2 - tam1) / 2
        snumero3 = snumero2[int(primerc):int(primerc+tam1)]
        x.append(float(str(0) + '.' +  snumero3))
        numero1 = int(snumero3)
    return x
```

```python
semilla = "5735"
data = cuadrados_medios(20, semilla)
print(data)
df = pd.DataFrame({'column':data}) 
df.to_csv('out_algoritmo_cuadrados_medios.csv', index=False, header=False,sep=';',decimal=',')
```
