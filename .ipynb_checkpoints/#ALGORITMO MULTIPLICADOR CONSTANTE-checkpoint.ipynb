```python
import numpy as np
import pandas  as pd
import os
import matplotlib.pyplot as plt
```

```python
def multiplicador_constante(n, semilla, constante):
    x = []
    tam1 = len(semilla)
    numero1 = int(semilla)
    for i in range(n):
        snumero = str(constante * numero1)
        tam2 = len(snumero)
        primerc = (tam2 - tam1) / 2
        snumero3 = snumero[int(primerc):int(primerc+tam1)]
        x.append(float(str(0) + '.' +  snumero3))
        numero1 = int(snumero3)
    return x
```

```python
semilla = "9803"
constante = 6965
data = multiplicador_constante(50, semilla, constante)
print(data)
df = pd.DataFrame({'column':data}) 
df.to_csv('out_algoritmo_multiplicador_constante.csv', index=False, header=False,sep=';',decimal=',')
```
