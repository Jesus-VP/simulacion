## ALGORITMOS DE PRODUCTOS MEDIOS - ESIS ( Nain Acero)

```python
import numpy as np
import pandas  as pd
import os
import matplotlib.pyplot as plt
```

```python
def productos_medios(n, semilla1, semilla2):
    x = []
    tam1 = len(semilla1)
    numero1 = int(semilla1)
    numero2 = int(semilla2)
    for i in range(n):
        snumero = str(numero1 * numero2)
        tam2 = len(snumero)
        primerc = (tam2 - tam1) / 2
        snumero3 = snumero[int(primerc):int(primerc+tam1)]
        x.append(float(str(0) + '.' +  snumero3))
        numero1 = numero2
        numero2 = int(snumero3)
    return x
```

```python
semilla1 = "5015"
semilla2 = "5734"
data = productos_medios(50, semilla1, semilla2)
print(data)
df = pd.DataFrame({'column':data}) 
df.to_csv('out_algoritmo_productos_medios.csv', index=False, header=['aleatorio'],sep=';',decimal='.')
df.to_csv('out_algoritmo_productos_medios.txt', index=False, header=['aleatorio'],sep=';',decimal='.')
```
